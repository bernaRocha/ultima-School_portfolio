Dentro das questões de humor decidi juntar as opções 7 e 9 e substituir por NaN como forma de facilitar a análise. Nas variáveis de raça/ etnia juntei as opções de outro com outro/ hispânico por entender se tratar de uma mesma categoria. Na variável de renda anual familiar decidi trocar as opções 12 (> que 20000), 13 (< que 20000), 77 - (Se recusou a responder) e 99 - (Não sabe) por NaN por entender que levaria a dubiedade.

```python
replace_map = { 
'DPQ010': {7: np.nan, 9: np.nan},
 'DPQ020': {7: np.nan, 9: np.nan},
 'DPQ030': {7: np.nan, 9: np.nan},
 'DPQ040': {7: np.nan, 9: np.nan},
 'DPQ050': {7: np.nan, 9: np.nan},
 'DPQ060': {7: np.nan, 9: np.nan},
 'DPQ070': {7: np.nan, 9: np.nan},
 'DPQ080': {7: np.nan, 9: np.nan},
 'DPQ090': {7: np.nan, 9: np.nan},
 'DMDEDUC': {7: np.nan, 9: np.nan},
 'RIDRETH1': {5: 4}, #colocando a categoria 5 na categoria 4
 'INDFMINC': {12: np.nan, # np.mean([20000, 75000])
             13: np.nan, # np.mean([0, 20000])
             77: np.nan,
             99: np.nan}
}
```

```python
df = df.replace(replace_map)
```

#### Criando a variável Q_somatorio com o somatório das questões de humor.

```python
Q = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9']
df["Q_somatorio"] = df[Q].sum(axis = "columns")
```

```python
df.Q_somatorio.value_counts()
```

### Classificando de acordo com a literatura <br>
0 (“sem sintomas”) se Q_somatorio < 5 <br>
1 (“sintomas leves”) se 5 <= Q_somatorio < 10 <br>
2 (“sintomas moderados”) se 10 <= Q_somatorio < 15 <br>
3 (“sintomas moderadamente severos”) se 15 <= Q_somatorio < 20 <br>
4 (“sintomas severos”) se Q_somatorio >= 20. <br>

```python
df["Q_somatorio"] = np.where(df["Q_somatorio"]< 5, 0, df["Q_somatorio"])
df["Q_somatorio"] = np.where(((df["Q_somatorio"] >= 5) & (df["Q_somatorio"] < 10)), 1, df["Q_somatorio"])
df["Q_somatorio"] = np.where(((df["Q_somatorio"] >= 10) & (df["Q_somatorio"] < 15)), 2, df["Q_somatorio"])
df["Q_somatorio"] = np.where(((df["Q_somatorio"] >= 15) & (df["Q_somatorio"] < 20)), 3, df["Q_somatorio"])
df["Q_somatorio"] = np.where(df["Q_somatorio"] >= 20, 4, df["Q_somatorio"])
```

```python
df.Q_somatorio.value_counts()
```

#### Resolvi juntar as categorias 3 e 4 na categoria 2 devido o número reduzido.

```python
df["Q_somatorio"] = df["Q_somatorio"].replace([3,4], 2)
```
