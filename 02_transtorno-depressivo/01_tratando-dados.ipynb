### Dicionário - DEMO_PHQ

<b>SEQN</b> - Identificador único do respondente

<b>DPQ010</b> - Nas últimas 2 semanas, com que frequência você ficou incomodado por algum dos problemas a seguir:

Pouco interesse ou pouco prazer em fazer as coisas <br>
**** Serve para todos os DPQ0 **** <br>
0 = Nenhuma vez <br>
1 = Menos de uma semana <br>
2 = Uma semana ou mais <br>
3 = Quase todos os dias <br>
7 = Se recusou a responder <br>
9 = Não sabe <br>

<b>DPQ020</b> - Nas últimas 2 semanas, com que frequência você ficou incomodado por algum dos problemas a seguir:

Se sentiu para baixo, deprimido(a) ou sem perspectiva

<b>DPQ030</b> - Nas últimas 2 semanas, com que frequência você ficou incomodado por algum dos 
problemas a seguir:

Dificuldade para pegar no sono ou permanecer dormindo ou dormiu mais do que o costume

<b>DPQ040</b> - Nas últimas 2 semanas, com que frequência você ficou incomodado por algum dos 
problemas a seguir:

Se sentiu cansado(a) ou com pouca energia

<b>DPQ050</b> - Nas últimas 2 semanas, com que frequência você ficou incomodado por algum dos 
problemas a seguir:

Falta de apetite ou comeu demais

<b>DPQ060</b> - Nas últimas 2 semanas, com que frequência você ficou incomodado por algum dos 
problemas a seguir:

Se sentiu mal consigo mesmo(a) ou achou que é um fracasso ou que decepcionou sua família ou a você mesmo(a)

<b>DPQ070</b> - Nas últimas 2 semanas, com que frequência você ficou incomodado por algum dos 
problemas a seguir:

Dificuldade para se concentrar nas coisas (como ler o jornal ou ver televisão)

<b>DPQ080</b> - Nas últimas 2 semanas, com que frequência você ficou incomodado por algum dos 
problemas a seguir:

Teve lentidão para se movimentar ou falar (a ponto de outras pessoas perceberem), ou ao contrário, esteve tão agitado(a) que você ficava andando de um lado para o outro mais do que costume

<b>DPQ090</b> - Nas últimas 2 semanas, com que frequência você ficou incomodado por algum dos 
problemas a seguir:

Pensou em se ferir de alguma maneira ou que seria melhor estar morto(a)

<b>RIAGENDR</b> - Gênero
1 = Masculino <br>
2 = Feminino

<b>RIDAGEYR</b> Idade em anos

<b>RIDRETH1</b> Raça/Etnia

7 = Se recusou a responder <br>
9 = Não sabe

<b>INDFMINC</b> Renda anual familiar (US$)

1 = 0-4999 <br> 
2 = 5000-9999 <br>
3 = 10000-14999 <br>
4 = 15000-19999 <br>
5 = 20000-24999 <br>
6 = 25000-34999 <br>
7 = 35000-44999 <br>
8 = 45000-54999 <br>
9 = 55000-64999 <br>
10 = 65000-74999 <br>

### Dicionário - PAG_HEI

<b>SEQN</b> - Identificador único do respondente

<b>PAG_MINW</b> - Tempo total semanal de atividade aeróbica moderada-vigorosa em minutos segundo o Physical Activity 
Guideline (PAG)

<b>ADHERENCE</b> - Grupo de aderência ao Physical Activity Guideline (PAG)

1 = Baixo (< 150 min/semana) <br>
2 = Adequado (150-300 min/semana) <br>
3 = Acima (> 300 min/semana) <br>

<b>HEI2015C1_TOTALVEG</b> Healthy Eating Index - Componente 1 - Total de Vegetais 0-5

<b>HEI2015C2_GREEN_AND_BEAN</b> Healthy Eating Index - Componente 2 - Verduras verde-escuras e leguminosas 0-5

<b>HEI2015C3_TOTALFRUIT</b> Healthy Eating Index - Componente 3 - Frutas totais 0-5

<b>HEI2015C4_WHOLEFRUIT</b> Healthy Eating Index - Componente 4 - Frutas in natura 0-5

<b>HEI2015C5_WHOLEGRAIN</b> Healthy Eating Index - Componente 5 - Grãos integrais 0-10

<b>HEI2015C6_TOTALDAIRY</b> Healthy Eating Index - Componente 6 - Laticínios 0-10

<b>HEI2015C7_TOTPROT</b> Healthy Eating Index - Componente 7 - Proteínas 0-5

<b>HEI2015C8_SEAPLANT_PROT</b> Healthy Eating Index - Componente 8 - Plantas marinhas e proteínas vegetais 0-5

<b>HEI2015C9_FATTYACID</b> Healthy Eating Index - Componente 9 - Ácido-graxo 0-10

<b>HEI2015C10_SODIUM</b> Healthy Eating Index - Componente 10 - Sódio 0-10

<b>HEI2015C11_REFINEDGRAIN</b> Healthy Eating Index - Componente 11 - Grãos refinados 0-10

<b>HEI2015C12_SFAT</b> Healthy Eating Index - Componente 12 - Gordura saturada 0-10

<b>HEI2015C13_ADDSUG</b> Healthy Eating Index - Componente 13 - Açúcares adicionados 0-10

<b>HEI2015_TOTAL_SCORE</b> Healthy Eating Index - Escore total 0-100

```python
#### Juntando os dataset's em um com apenas os adultos
df = pd.merge(df_adultos, df_crianca_adulto, how = 'left', on = 'SEQN')
```

```python
df.shape
```

```python
df
```

```python
df.info()
```

```python
df = df.drop('SEQN', 1)
```

```python
df
```

```python
#### Renomeando as colunas
df = df.rename(columns={'DPQ010': 'Q1',
                        'DPQ020': 'Q2',
                        'DPQ030': 'Q3',
                        'DPQ040': 'Q4',
                        'DPQ050': 'Q5',
                        'DPQ060': 'Q6',
                        'DPQ070': 'Q7',
                        'DPQ080': 'Q8',
                        'DPQ090': 'Q9',
                        'RIAGENDR': 'genero',
                        'RIDAGEYR': 'idade',
                        'RIDRETH1': 'raca',
                        'DMDEDUC': 'escolaridade',
                        'INDFMINC': 'renda_anual_familiar',
                        'PAG_MINW': 'tempo_atividade_semanal_pag',
                        'ADHERENCE': 'aderencia_pag',
                        'HEI2015C1_TOTALVEG': 'total_vegetais',
                        'HEI2015C2_GREEN_AND_BEAN': 'total_verdura-v-escura-leguminosas',
                        'HEI2015C3_TOTALFRUIT': 'total_frutas',
                        'HEI2015C4_WHOLEFRUIT': 'total_frutas_in-natura',
                        'HEI2015C5_WHOLEGRAIN': 'total_graos-integrais',
                        'HEI2015C6_TOTALDAIRY': 'total_laticinios',
                        'HEI2015C7_TOTPROT': 'total_proteinas',
                        'HEI2015C8_SEAPLANT_PROT': 'total_planta-marinha-proteina-vegetal',
                        'HEI2015C9_FATTYACID': 'total_acido-graxo',
                        'HEI2015C10_SODIUM': 'total_sodio',
                        'HEI2015C11_REFINEDGRAIN': 'total_graos-refinados',
                        'HEI2015C12_SFAT': 'total_gordura-saturada',
                        'HEI2015C13_ADDSUG': 'total_acucar-adicionado',
                        'HEI2015_TOTAL_SCORE': 'score_saude'})
```

#### Renomeando registros (Linhas)

```python
df.genero = df.genero.replace({1: 'homem', 2: 'mulher'})
```

```python
df.raca = df.raca.replace({1: 'Branco nao-hispanico', 
                           2: 'Negro nao-hispanico ',
                           3: 'Mexicano-americano',
                           4: 'Outro',
                           5: 'Outro-Hispanico '})
```

```python
df.escolaridade = df.escolaridade.replace({1: 'Menor que 9o ano', 
                                           2: '9-12o ano', 
                                           3: 'Ensino medio', 
                                           4:'Ensino superior incompleto', 
                                           5:'Superior completo ou maior', 
                                           7: 'Se recusou a responder',
                                           9: 'Nao sabe'})
```

```python
df.aderencia_pag = df.aderencia_pag.replace({1.0: 'Baixo (< 150 min/semana)', 
                                             2.0: 'Adequado (150-300 min/semana)', 
                                             3.0: 'Acima (> 300 min/semana)'})
```

```python
### criando o csv com os dados tratados
df.to_csv('data/dados_tratados.csv', index=False)
```
